```python
store_revenue = df.groupby('store_location')['revenue'].sum()
print(store_revenue) # revenue calculated for the store locations
```

```python
# plot store-wise revenue using bar chart
ax=store_revenue.plot(kind='bar',figsize=(6,5))
plt.title('Store-wise Revenue')
plt.xlabel('Store Location')
plt.ylabel('Revenue')
for container in ax.containers:
  ax.bar_label(container,fmt='%.2f')
plt.show()
```

```python
print(store_revenue.idxmax(), store_revenue.max())
```

```python
#store-wise quantity sold and revenue earned
store_wise_revenue = df.groupby('store_location')[['transaction_qty','revenue',]].sum()
print(store_wise_revenue)
```

```python
#plot store-wise quantity sold and earned revenue
store_summary = store_wise_revenue.plot(kind='bar',color=['red','Green'])
plt.title('Store-wise Quantity sold vs Revenue Earned')
plt.xlabel('Store location')
plt.ylabel('Value')
plt.xticks(rotation=0)
for container in store_summary.containers:
  store_summary.bar_label(container,fmt='%.2f')
plt.show()
```

```python
category_sales = df.groupby('product_category')['transaction_qty'].sum()
print(category_sales) # calculating transaction_qty sold for each category
```

```python
#plot category_sales using bar chart
c_sales = category_sales.plot(kind='bar',figsize=(6,5))
plt.title('Product Category Sales')
plt.xlabel('Category')
plt.ylabel('Quantity Sold')
for container in c_sales.containers:
  c_sales.bar_label(container)
plt.show()
```

```python
print(category_sales.idxmax(),category_sales.max())
```

```python
#store sales
store_sales = df.groupby('store_location')['transaction_qty'].sum()
print(store_sales)
```

```python
#plot store sales store-wise quantity sold
s_sales = store_sales.plot(kind='bar',figsize=(6,5))
plt.title('Store Sales')
plt.xlabel('Store_location')
plt.ylabel('Quantity sold')
for container in s_sales.containers:
  s_sales.bar_label(container)
plt.show()
```

```python
# hourly sales
hourly_sales = df.groupby('hour')['transaction_qty'].sum()
print(hourly_sales)
```

```python
hourly_sales.plot(kind='bar')
plt.title('hourly sales')
plt.xlabel('Hour')
plt.ylabel('Quanity sold')
plt.show()
```

```python
#retrive busiest hour and what product sold and in which store location
busiest_hour = hourly_sales.idxmax()
busiest_data = df[df['hour']==busiest_hour]
result = busiest_data.groupby(['store_location','product_category'])['transaction_qty'].sum()
top_result = result.idxmax()
quantity= result.max()
print('Busiest hour:',f"{busiest_hour}:00")
print("Location:",top_result[0])
print('Top category:',top_result[1])
print('Quantity sold:',quantity)
```

```python
# top product sold w.r.t store location and hour based
result = df.groupby(['hour','store_location','product_category'])['transaction_qty'].sum().reset_index()
top_per_store_hour = result.loc[result.groupby(['hour','store_location'])['transaction_qty'].idxmax()]
print(top_per_store_hour)
```

```python
#best selling product in each store
#Quantity sold by store and product category
category_sales_store = df.groupby(['store_location','product_category'])['transaction_qty'].sum().reset_index()
#get top category in each store
top_category = category_sales_store.loc[category_sales_store.groupby('store_location')['transaction_qty'].idxmax()]
print(top_category)
```

```python
#plot category_sales_store based using horizantal bar chart
# top category is coffee
top_c = top_category.plot(x='store_location',
                  y='transaction_qty',kind='barh',color='skyblue',legend=False)
plt.xlabel('Quantity Sold')
plt.ylabel('Store location')
plt.title('Top category sold in each store')
for container in top_c.containers:
  top_c.bar_label(container)
plt.show()
```

```python
#plot coffee sales across stores for every hour using Grouped bar chart
pivot = top_per_store_hour.pivot(index='hour',columns='store_location',values='transaction_qty')
coffee_sales = pivot.plot(kind='bar',figsize=(12,8),color=['Steelblue','darkorange','green'])
plt.title('Top category(coffee) sales by hour and store')
plt.xlabel('Hour')
plt.ylabel('Quantity sold')
plt.xticks(rotation=0)
plt.legend(title='Store')
for container in coffee_sales.containers:
  coffee_sales.bar_label(container)
plt.show()
```
