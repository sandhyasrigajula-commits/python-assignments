```python
plt.figure(figsize=(10,5))
sns.scatterplot(data=df, x = 'age',y='salary', color ='blue',edgecolor='black',alpha=0.5)
plt.title('scatterplot of employee age vs employee salary')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='healthy_eating', y='active_lifestyle', color='red', edgecolor='black', alpha=0.5)
plt.title('Scatterplot of healthy eating vs active lifestyle')
plt.xlabel('Healthy eating')
plt.ylabel('Active lifestyle')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='healthy_eating', y='salary', color='b', edgecolor='black', alpha=0.5)
plt.title('Scatterplot of Healthy eating vs Salary')
plt.xlabel('healthy eating')
plt.ylabel('Salary')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='groups', edgecolor='black', alpha=0.7)
plt.title('Countplot of groups')
plt.xlabel('Groups')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(data=df, x='salary', color='orange', edgecolor='black', alpha=0.7, bins=10)
plt.title('Histogram of salary')
plt.xlabel('Salary')
plt.ylabel('Count')
plt.show()
```
