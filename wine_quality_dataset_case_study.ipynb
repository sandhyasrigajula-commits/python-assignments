```python
plt.figure(figsize = (11,6))

sns.histplot(data = df, x = 'fixed acidity', color = 'RED', alpha = 0.5, bins = 5)

plt.title('Histogram For Fixed Acidity')
plt.xlabel('Fixed Acidity')
plt.ylabel('Count')
plt.show()
```

```python
print(df['fixed acidity'].mean())
```

```python
print(df['fixed acidity'].median())
```

```python
plt.figure(figsize=(11,6))

sns.histplot(data = df, x = 'fixed acidity', color = 'green',alpha = 0.5, edgecolor= 'linen', bins = 5)
plt.xlabel('Fixed Acidity')
plt.ylabel('Count')
plt.vlines(df['fixed acidity'].mean(), ymin = 0, ymax = 4000, colors = 'blue', label = 'Mean')
plt.vlines(df['fixed acidity'].median(), ymin= 0, ymax = 4000, colors = 'red', label = 'Median')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(11,6))

sns.histplot(data = df, x = 'volatile acidity', color = 'orange', alpha = 0.5, bins = 5)

plt.title('Histogram For Volatile Acidity')
plt.xlabel('Volatile Acidity')
plt.ylabel('Count')
plt.show()
```

```python
df['volatile acidity'].skew()
```

```python
plt.figure(figsize=(11,6))

sns.distplot(df['volatile acidity'], color = 'pink')

plt.title('Distplot For Volatile Acidity')
plt.xlabel('Volatile Acidity')
plt.ylabel('Density')
plt.show()
```

```python
print(df['volatile acidity'].skew())
print(df['volatile acidity'].mean())
print(df['volatile acidity'].median())
```

```python
plt.figure(figsize=(11,6))

sns.histplot(data = df, x = 'volatile acidity', color = 'orange',alpha = 0.5, edgecolor= 'linen', bins = 5)
plt.xlabel('Volatile Acidity')
plt.ylabel('Density')
plt.vlines(df['volatile acidity'].mean(), ymin = 0, ymax = 4000, colors = 'blue', label = 'Mean')
plt.vlines(df['volatile acidity'].median(), ymin= 0, ymax = 4000, colors = 'red', label = 'Median')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(11,6))

sns.histplot(data = df, x = 'citric acid', color = 'orange', alpha = 0.5, bins = 5)

plt.title('Histogram For Citric Acid')
plt.xlabel('Aitric Acid')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(11,6))

sns.histplot(data = df, x = 'citric acid', color = 'green',alpha = 0.5, edgecolor= 'linen', bins = 5)
plt.xlabel('citric acidy')
plt.ylabel('Density')
plt.vlines(df['citric acid'].mean(), ymin = 0, ymax = 4000, colors = 'blue', label = 'Mean')
plt.vlines(df['citric acid'].median(), ymin= 0, ymax = 4000, colors = 'red', label = 'Median')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(11,6))

sns.distplot(df['citric acid'], color = 'pink')

plt.title('Distplot For Citric Acid')
plt.xlabel('Citric Acid')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(11,6))

sns.countplot(x='quality', data=df)
plt.xlabel('Quality')
plt.ylabel('Count')
plt.show()
```
